```python
import pandas as pd
df = pd.read_csv("autos_prepared.csv")
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(df["kilometer"], df["price"])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(df[["kilometer"]], df[["price"]])

print("Intercept: " +str(model.intercept_))
print(" Coef: "+str(model.coef_))
```

```python
min_x = min(df["kilometer"])
max_x = max(df["kilometer"])

predichted = model.predict([[min_x], [max_x]])
predichted2 = model.predict([[50000]])
print(predichted2)
```

```python
plt.scatter(df["kilometer"], df["price"])
plt.plot([min_x,max_x], predichted, color="red")
plt.show
```
